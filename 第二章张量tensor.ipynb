```python
tensor_2d=torch.tensor([
    [-9,4,2,4,5],
    [2,3,5,6,7],
    [2,4,5,6,7],
    [6,4,2,1,9]
])
print(tensor_2d)
print(tensor_2d.shape)
```

```python
tensor_3d=torch.stack([tensor_2d,tensor_2d+10,tensor_2d-5])
print(tensor_3d)
print(tensor_3d.shape)
```

```python
tensor_4d=torch.stack([tensor_3d,tensor_3d+100])
print(tensor_4d)
print(tensor_4d.shape)
```

```python
# 创建 5D 张量（矩阵的立方体）
tensor_5d = torch.stack([tensor_4d, tensor_4d + 1000])  # 堆叠 2 个 4D 张量
print("5D Tensor (Matrix of Cubes):\n", tensor_5d)
print("Shape:", tensor_5d.shape)  # 形状
```

### 形状变换操作

```python
reshaped=t.view(3,2)#这个必须元素个数相同，不能view4，2这样元素个数不对
print(reshaped)
```

```python
flattened=t.flatten()
print(flattened)
```

### 数学运算操作

```python
t_add=t+10
t_add
```

```python
t_mul=t*2
t_mul
```

```python
t_sum=t.sum()
t_sum
```

```python
t2=torch.tensor([[1,1,1],[1,1,1]],dtype=torch.float32)
print(t)
print(t2)
tensor_dot=torch.matmul(t,t2.T)
tensor_dot
```
